## Number of incidents per states per year??

## Which month wife's have more tendency to hurt husband??

```python
Wife_husband_month = homicide[['Month','Relationship','Victim Sex','Perpetrator Sex','Record ID']]
Wife_husband_month = Wife_husband_month[(Wife_husband_month['Relationship']=='Husband') & 
                                       (Wife_husband_month['Victim Sex']=='Male') &
                                        (Wife_husband_month['Perpetrator Sex']=='Female')]
Wife_husband_month = Wife_husband_month.groupby(['Month'])['Record ID'].count().reset_index()
Wife_husband_month = Wife_husband_month.sort_values(by='Record ID',ascending = False)
Wife_husband_month
```

## Number of Incident involving Ex-husband vs Ex-wife??

```python
Exhusband_Exwife = homicide[['Relationship','Record ID','Victim Sex','Perpetrator Sex']]
Exhusband_Exwife = Exhusband_Exwife[((Exhusband_Exwife['Relationship'] == 'Ex-Wife') &
                                    (Exhusband_Exwife['Victim Sex']=='Female') & 
                                     (Exhusband_Exwife['Perpetrator Sex']=='Male')) | 
                                    ((Exhusband_Exwife['Relationship'] == 'Ex-Husband') &
                                     (Exhusband_Exwife['Victim Sex']=='Male') & 
                                     (Exhusband_Exwife['Perpetrator Sex']=='Female'))]
Exhusband_Exwife = Exhusband_Exwife.groupby('Relationship')['Record ID'].count().reset_index()
Exhusband_Exwife = Exhusband_Exwife.sort_values(by = 'Record ID',ascending = False)
Exhusband_Exwife
```

## Number of Incident involving Exployee vs employer??

```python
Employee_Employer = homicide[['Relationship','Record ID']]
Employee_Employer = Employee_Employer[(Employee_Employer['Relationship'] == 'Employer') | 
                                    (Employee_Employer['Relationship'] == 'Employee')]
Employee_Employer = Employee_Employer.groupby('Relationship')['Record ID'].count().reset_index()
Employee_Employer = Employee_Employer.sort_values(by = 'Record ID',ascending = False)
Employee_Employer
```

```python
homicide['Family'] = 0
for i in range(len(homicide)):
    if (homicide.loc[i,'Relationship'] == 'Brother') & (homicide.loc[i,'Perpetrator Sex'] == 'Male'):
        homicide.loc[i,'Family'] = 'Brother Killed by Brother'
        
    elif (homicide.loc[i,'Relationship'] == 'Brother') & (homicide.loc[i,'Perpetrator Sex'] == 'Female'):
        homicide.loc[i,'Family'] = 'Brother Killed by Sister'
        
    elif (homicide.loc[i,'Relationship'] == 'Sister') & (homicide.loc[i,'Perpetrator Sex'] == 'Female'):
        homicide.loc[i,'Family'] = 'Sister Killed by Sister'
        
    elif (homicide.loc[i,'Relationship'] == 'Sister') & (homicide.loc[i,'Perpetrator Sex'] == 'Male'):
        homicide.loc[i,'Family'] = 'Sister Killed by Brother'
        
    elif (homicide.loc[i,'Relationship'] == 'Mother') & (homicide.loc[i,'Perpetrator Sex'] == 'Male'):
        homicide.loc[i,'Family'] = 'Mother Killed by son'
        
    elif (homicide.loc[i,'Relationship'] == 'Mother') & (homicide.loc[i,'Perpetrator Sex'] == 'Female'):
        homicide.loc[i,'Family'] = 'Mother Killed by Daugher'
        
    elif (homicide.loc[i,'Relationship'] == 'Son') & (homicide.loc[i,'Perpetrator Sex'] == 'Female'):
        homicide.loc[i,'Family'] = 'Son Killed by Mother'
        
    elif (homicide.loc[i,'Relationship'] == 'Son') & (homicide.loc[i,'Perpetrator Sex'] == 'Male'):
        homicide.loc[i,'Family'] = 'Son Killed by Father'
        
    elif (homicide.loc[i,'Relationship'] == 'Daughter') & (homicide.loc[i,'Perpetrator Sex'] == 'Female'):
        homicide.loc[i,'Family'] = 'Daughter Killed by Mother'
        
    elif (homicide.loc[i,'Relationship'] == 'Daughter') & (homicide.loc[i,'Perpetrator Sex'] == 'Male'):
        homicide.loc[i,'Family'] = 'Daughter Killed by Father'
        
    elif (homicide.loc[i,'Relationship'] == 'Wife') & (homicide.loc[i,'Perpetrator Sex'] == 'Male'):
        homicide.loc[i,'Family'] = 'Wife Killed by Husband'
        
    elif (homicide.loc[i,'Relationship'] == 'Husband') & (homicide.loc[i,'Perpetrator Sex'] == 'Female'):
        homicide.loc[i,'Family'] = 'Husband Killed by Wife'
        
    elif (homicide.loc[i,'Relationship'] == 'Father') & (homicide.loc[i,'Perpetrator Sex'] == 'Male'):
        homicide.loc[i,'Family'] = 'Father Killed by Son'
        
    elif (homicide.loc[i,'Relationship'] == 'Father') & (homicide.loc[i,'Perpetrator Sex'] == 'Female'):
        homicide.loc[i,'Family'] = 'Father Killed by Daughter'
        
    else:
        homicide.loc[i,'Family'] = 'Others'
        
homicide.head()
```

### Create the DTM with min_df = .02 and max_df = 0.90 identify the top 3 words

```python
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from matplotlib import pyplot as plt
```

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
words = set(stopwords.words('english'))
words
```

```python
amazon = pd.read_excel("D:/EDA/Workspace/Amazon.xlsx")
text = amazon[['reviews.text']]
print(text['reviews.text'][0])
text.head()
```

```python
## Step-1 & Step-2 & Step-3 : Converting everything into lower case and replacing the special charater

text['clear_text'] = text['reviews.text'].str.lower().str.replace('[^a-z ]','')
text.head()
```

```python
## Step-4 : Removing the stopwords

stop = stopwords.words('english')

def sw(text):
    text = [word for word in text.split() if word not in stop]
    return " ".join(text)

## Applying the sw on clear_text columns

text['text_without_stop'] = text['clear_text'].apply(sw)
print(text['clear_text'][0])
print(text['text_without_stop'][0])
```

```python
## Starting the count vector

vec = CountVectorizer()

vec.fit(text['text_without_stop'])
```

```python
## DTM Matrix creation

x = vec.fit_transform(text['text_without_stop'])
x
```

```python
## Convert this DTM into dataframe

DTM = pd.DataFrame(x.toarray(), columns = vec.get_feature_names())

DTM.head()
```

```python
# Calculating the word freq

word_freq = DTM.sum()
word_freq.sort_values(ascending = False).head(3)
```
