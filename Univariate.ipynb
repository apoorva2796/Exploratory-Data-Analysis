```python
Hours_Tweets = Modi.groupby(['Hours'])['id'].count().reset_index()
Hours_Tweets.sort_values(by='id',ascending = False)
```

```python
Source_week_Tweets = Modi.groupby(['source','Week_Day'])['id'].count().reset_index()
Source_week_Tweets.sort_values(by='id',ascending = False)
```

## Univariate analysis - Text Column

```python
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from matplotlib import pyplot as plt
```

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
words = set(stopwords.words('english'))
words
```

```python
Modi.shape
```

```python
text = Modi[['text']]
print(text['text'][0])
text.head()
```

```python
# Number of words in each row

text['count'] = text['text'].apply(lambda x: len(str(x).split(" ")))
text.head()
```

```python
## Step-1 & Step-2 & Step-3 : Converting everything into lower case and replacing the special charater

text['clear_text'] = text['text'].str.lower().str.replace('[^a-z ]','')
text.head()
```

```python
## Step-4 : Removing the stopwords

stop = stopwords.words('english')
stop.extend(['amp','modi','narendra','pmo'])

def sw(text):
    text = [word for word in text.split() if word not in stop]
    return " ".join(text)

## Applying the sw on clear_text columns

text['text_without_stop'] = text['clear_text'].apply(sw)
print(text['clear_text'][0])
print(text['text_without_stop'][0])
```

```python
## Starting the count vector

vec = CountVectorizer()

vec.fit(text['text_without_stop'])

# Creating the dict. of item in this matrix

dic = vec.vocabulary_.items()
dic
```

```python
## DTM Matrix creation

x = vec.fit_transform(text['text_without_stop'])
x
```

```python
## Convert this DTM into dataframe

DTM = pd.DataFrame(x.toarray(), columns = vec.get_feature_names())

DTM.head()
```

```python
# Calculating the word freq

word_freq = DTM.sum()
word_freq.sort_values(ascending = False).head()
```
